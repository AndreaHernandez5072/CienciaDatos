```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report,confusion_matrix,r2_score,mean_squared_error,mean_absolute_error
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv("weatherHistory.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df["Formatted Date"]=df["Formatted Date"].apply(lambda x:x[:13])
df["Formatted Date"]=pd.to_datetime(df["Formatted Date"])
```

```python
def display_mising_value(df):
  for col in df.columns.tolist():
    print("{} column missing values: {}".format(col,df[col].isnull().sum()),".(",(df[col].isnull().sum()/df.shape[0])*100,"%)")
    print("\n")

display_mising_value(df)
```

```python
df["Precip Type"]= df["Precip Type"].fillna("no rain")
```

```python
df["Year"]=df["Formatted Date"].dt.year
df["Month"]=df["Formatted Date"].dt.month
df["Day"]=df["Formatted Date"].dt.day
df["Hour"]=df["Formatted Date"].dt.hour
```

```python
df.head()
```

```python
def FrecuencyEncoding(dataset,columna):
  frecuency=dataset[columna].value_counts(normalize=True)
  dataset[columna] = dataset[columna].map(frecuency)
  return dataset[columna]

df["Summary"] = FrecuencyEncoding(df,"Summary")
df["Precip Type"]=FrecuencyEncoding(df,"Precip Type")
df["Daily Summary"]=FrecuencyEncoding(df,"Daily Summary")
```

```python
df = df.drop( columns = ["Formatted Date"])
```

```python
df["Loud Cover"].unique()
```

```python
df=df.drop(columns=["Loud Cover"])
```

```python
df
```

```python
def CuantificaOutliers (dataset):
  for col in dataset.columns:
    q1, q3 = np.percentile(dataset[col], [25, 75])
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    outlier = dataset[(dataset[col] < lower_bound)
                  | (dataset[col] > upper_bound)]
    print(col,":  (",outlier.shape[0]," / ",dataset.shape[0]," )  ",(outlier.shape[0] /  dataset.shape[0])*100,"%")

#GraficaOutliers(df)
CuantificaOutliers(df)
```

```python
corr=df.corr()
round(corr,2)
```

```python
df.to_csv("weatherHistoriyClean.csv",index=False)
```

```python
df2=df
y=df2["Temperature (C)"]
x=df2.drop(columns=["Wind Speed (km/h)","Wind Bearing (degrees)","Pressure (millibars)","Year","Day","Temperature (C)"])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=199)
```

```python
#Entrenar modelo lineal
lr = LinearRegression()
lr.fit(x_train,y_train)
#Predecir en base al modelo entrenado
y_hat=lr.predict(x_test)
y_hat
show="-"*17
#Evaluanod modelo
print(f"{show}Resultados{show} \n")
#Usa el simulador para evaluar el error es mean_squared_error
print("Mean square error: ", mean_squared_error(y_test,y_hat))
print("Mean absolute error: ", mean_absolute_error(y_test,y_hat))
print("Mean square error: ", r2_score(y_test,y_hat))
```
